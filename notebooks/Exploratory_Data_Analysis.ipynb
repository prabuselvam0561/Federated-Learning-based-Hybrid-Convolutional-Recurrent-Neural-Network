# Exploratory Data Analysis (EDA)

This notebook performs an exploratory data analysis of the IoT dataset. The goal is to understand the data, identify patterns, and prepare it for model training.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
dataset_path = '../datasets/sample_data/training_set.csv'
data = pd.read_csv(dataset_path)

# Display basic information
print(data.info())
print(data.head())
```

```
# Visualize class distribution
plt.figure(figsize=(10, 6))
sns.countplot(data['label'])
plt.title('Class Distribution')
plt.xticks(rotation=45)
plt.show()
```
